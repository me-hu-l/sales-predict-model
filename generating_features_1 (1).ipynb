```python
grid = []
cols = ['shop_id', 'item_id', 'date_block_num']

for month in range(34):
  curr_items = sales.loc[sales.date_block_num==month, 'item_id'].unique()
  curr_shops = sales.loc[sales.date_block_num==month, 'shop_id'].unique()
  grid.append(np.array(list(product(curr_items, curr_shops, [month])), dtype= np.int32))
grid = pd.DataFrame(np.concatenate(grid,axis=0), columns=cols,dtype = np.int32)
```

```python
sales_with_categories = sales.merge(items,on=['item_id'],how='left')
gb1 = sales_with_categories.drop(['shop_id', 'item_price','item_id'],axis=1).groupby(['item_category_id', 'date_block_num'],as_index=False).sum()
gb1 = gb1.rename(columns={'item_cnt_day':'item_cat_block_target_sum'})
gb2 = sales_with_categories.drop(['shop_id', 'item_price','item_id'],axis=1).groupby(['item_category_id', 'date_block_num'],as_index=False).mean()
gb2 = gb2.rename(columns={'item_cnt_day': 'item_cat_block_target_mean'})
```

```python
gb = gb1.merge(gb2,on=['item_category_id','date_block_num'],how='left')
gb
```

```python
all_data = all_data.merge(gb, on=['item_category_id','date_block_num'],how='left').fillna(0)
```

```python
#all_data = downcast_dtypes(all_data)
del grid,gb,gb1,gb2, sales_with_categories
```

```python
index_cols = ['shop_id', 'item_id', 'date_block_num','item_category_id']
cols_to_rename = list(all_data.columns.difference(index_cols))
for i in ['item_name']:
    cols_to_rename.remove(i)
print(cols_to_rename)
cols_gb_item = [i for i in cols_to_rename if 'item_block' in i]
cols_gb_shop = [i for i in cols_to_rename if 'shop_block' in i]
cols_gb_cat = [i for i in cols_to_rename if 'item_cat' in i]
cols_gb_all = ['target']
cols_gb_key=[['item_id'],['shop_id'],['item_category_id'],['shop_id','item_id']]
cols_gb_value = [cols_gb_item,cols_gb_shop,cols_gb_cat,cols_gb_all]
print(cols_gb_value)
```

```python
del item_categories,items,sales,shops,test
```

```python
shift_range = [1,2,3]
i=1
for month_shift in shift_range:
    for k,v in zip(cols_gb_key,cols_gb_value): 
        index_col = ['date_block_num'] + k
        train_shift = all_data[index_col + v].copy().drop_duplicates()

        train_shift['date_block_num'] = train_shift['date_block_num'] + month_shift

        foo = lambda x: '{}_lag_{}'.format(x, month_shift) if x in v else x
        train_shift = train_shift.rename(columns=foo)
        all_data = pd.merge(all_data, train_shift, on=index_col, how='left').fillna(0)
        print(i)
        i+=1
```
